<font color=#B94CE1>MODEL TRAINING</font>

```python
array = fetal.values
random.shuffle(array)
random.shuffle(array)
X = array[:,0:21]
Y = array[:,21]

validation_size = 0.2
seed = 101
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size = validation_size, random_state = seed)
```

```python
mod = DecisionTreeClassifier() 
mod.fit(X_train,Y_train)
pred = mod.predict(X_test)
pred
```

<font color=#B94CE1>REPORT</font>

```python
report = classification_report(Y_test,pred)
print(report)
```

```python
preds = pd.DataFrame({'y_pred': pred, 'y_test':Y_test})
preds = preds.sort_values(by='y_test')
preds = preds.reset_index()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(preds['y_pred'], label='pred', color="pink")
plt.plot(preds['y_test'], label='actual')
plt.legend()
plt.show()
```
